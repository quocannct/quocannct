```python
baseurl = 'https://5theway.com'
productlinks = set()  

for x in range(1, 4):
    url = f'{baseurl}/collections/all?page={x}'
    r = requests.get(url)

    if r.status_code == 200:
        soup = BeautifulSoup(r.text, 'lxml')
        productlist = soup.find_all('div', class_='product-block')

        for item in productlist:
            for link in item.find_all('a', href=True):
                href = link['href']
                full_url = f'{baseurl}{href}' 
                if full_url not in productlinks:
                    print(full_url)
                    productlinks.add(full_url)
    else:
        print(f"Failed to retrieve page {x}. Status code: {r.status_code}")
num_products = len(productlinks)
print(f"Number of unique products: {num_products}")
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

baseurl = 'https://5theway.com'

driver = webdriver.Chrome()

product_data = []

for x in range(1, 4):
    url = f'{baseurl}/collections/all?page={x}'
    driver.get(url)

    driver.implicitly_wait(10)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    productlist = soup.find_all('div', class_='product-block')

    for item in productlist:
        product_info = {}

        product_name_element = item.find('h3', class_='pro-name')
        if product_name_element:
            product_info['Product Name'] = product_name_element.text.strip()

        product_link_element = item.find('a', href=True)
        if product_link_element:
            product_info['Product Link'] = f"{baseurl}{product_link_element['href']}"

        product_price_element = item.find('p', class_='pro-price')
        if product_price_element:
            product_info['Product Price'] = product_price_element.text.strip()

        product_color_elements = item.find_all('span', class_='variant-image-loop')
        if product_color_elements:
            product_info['Product Colors'] = [color['title'] for color in product_color_elements]

        # Use Selenium to retrieve image URLs
        image_elements = item.find_all('img', class_='img-loop')
        if image_elements:
            image_urls = [img['src'] for img in image_elements]
            product_info['Image URLs'] = image_urls

        product_data.append(product_info)
driver.quit()

# Print result
for i, product_info in enumerate(product_data, start=1):
    print(f"Product {i}:")
    print("Product Name:", product_info.get('Product Name', 'N/A'))
    print("Product Link:", product_info.get('Product Link', 'N/A'))
    print("Product Price:", product_info.get('Product Price', 'N/A'))
    print("Product Colors:", product_info.get('Product Colors', 'N/A'))
    print("Image URLs:", product_info.get('Image URLs', 'N/A'))
    print("\n")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import Row

spark = SparkSession.builder.appName("ProductData").getOrCreate()

schema = ["Product Number", "Product Name", "Product Link", "Product Price", "Product Colors", "Image URLs"]

product_data_rows = []
for product_number, product in enumerate(product_data, start=1):
    product_name = product.get("Product Name", "N/A")
    product_link = product.get("Product Link", "N/A")
    product_price = product.get("Product Price", "N/A")
    product_colors = product.get("Product Colors", ["N/A"])
    
    image_urls = product.get("Image URLs", ["N/A"])
    
    product_data_rows.append(Row(product_number, product_name, product_link, product_price, product_colors, image_urls))

product_df = spark.createDataFrame(product_data_rows, schema=schema)

product_df.show(n=1000)
# product_df.show(n=1000, truncate=False)
spark.stop()
```
